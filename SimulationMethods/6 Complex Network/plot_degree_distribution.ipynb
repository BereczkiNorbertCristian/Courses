```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 100
```

```python
f = open("degree.txt","r")
degree = [0] * N
for i in range(N):
    degree[i] = int(f.readline())
```

```python
degree[:10]
```

```python
ser = pd.Series(data=degree)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,6))
sns.distplot(ser,ax=ax)
```
